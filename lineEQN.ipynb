```python
import random as rand
import matplotlib.pyplot as plt
def lineEqn(m,x,c):
    return(m*x+c)
randomX=[]
randomY=[]
for i in range(0,1000):
    randomX.append(rand.randint(0,1000))
for i in range(0,len(randomX)):
    randomY.append(lineEqn(5,randomX[i],4))
plt.scatter(randomX,randomY)
plt.show()
```
