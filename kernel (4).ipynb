```python
import random as rand
import matplotlib.pyplot as plt
def parabola(x):
    return(x*x)
randomX=[]
randomY=[]
for i in range(-1000,1000):
    randomX.append(rand.randint(-1000,1000))
for i in range(0,len(randomX)):
    randomY.append(parabola(randomX[i]))
plt.plot(randomX,randomY)
plt.show()
```
