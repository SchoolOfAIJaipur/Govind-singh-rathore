```python
import random as rand
import matplotlib.pyplot as plt
randomX=[]
randomY=[]
def parabola(a,x):
    return((4*a*x)**0.5)
for i in range(0,1000):
    randomX.append(rand.randint(0,1000))
for i in range(0,1000):
    randomY.append(parabola(2,randomX[i]))
plt.title('my first plot')
plt.plot(randomX,randomY)
plt.show()
```
